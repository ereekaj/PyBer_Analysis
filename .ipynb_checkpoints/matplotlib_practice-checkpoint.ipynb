```python
%matplotlib inline

#Import Dependencies
import matplotlib.pyplot as plt
```

```python
savings= [0, 100, 150, 200, 300, 500, 300, 350, 500, 600, 750, 1000]
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 
         'Dec']
```

```python
plt.plot(months, savings)
plt.xlabel('Month')
plt.ylabel('Savings ($)')
plt.title('Savings Account Results')
plt.show()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create the plot
plt.plot(x_axis, y_axis)
plt.xlabel("Months")
plt.ylabel('Total Fare')
plt.title('Total Fare Per Month')
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
#add gridlines
plt.grid()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# Create the plot
# Create the plot.
plt.barh(x_axis, y_axis, color="magenta", label='Boston')
#change yaxis to ascending order
plt.gca().invert_yaxis()
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
#Scatterplots
x = [50, 42, 39, 34, 32, 48, 22, 17, 8, 13]
y = [17, 19, 21, 22, 24, 23, 34, 31, 43, 35]
```

```python
plt.scatter(x, y, c='cyan', marker = '*', s=x)
plt.xlabel('Total Number of Rides (Per City)')
plt.ylabel('Average Fare ($)')
plt.title('Average Far vs Total Rides')
plt.show()
```

```python
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

plt.scatter(y_axis, x_axis, c ='#89cff0', s = [i * 5 for i in y_axis], alpha = 0.8, marker = 'o', edgecolors='black', label = "Chicago")
plt.gca().invert_yaxis()
plt.xlabel('Date')
plt.ylabel('Fare ($)')
plt.title('PyBer Fare by Month')
plt.legend()
plt.xlim(0,50)
plt.show
```

```python
#create pie chart
categories = ['pants', 'shirts', 'outerwear', 'pajamas', 'underwear', 'shoes']
revenue = [25, 15, 5, 20, 10, 25]
explode = [0, .1, 0, 0, 0, 0]
```

```python
plt.pie(revenue, labels = categories, autopct="%1.1f%%", explode = explode)
plt.show()
```

```python
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    shadow = True, 
    startangle = 85,
    labels=x_axis,
    autopct='%.2f%%')

plt.show()
```
